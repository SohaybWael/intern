```python
#importing libraries for data pre-processing
import numpy as np
import pandas as pd
```

```python
#loading dataset
data=pd.read_csv("Datasets/Employee.csv")
data.head()
```

```python
#creating independent and dependent vector variables
independent=data.iloc[:,:-1].values
dependent=data.iloc[:,-1].values
```

```python
#handling missing values

#sum of missing data in the dataset
print(data.isnull().sum())

#first approach: handling the missing values by dropping the records that contain the missing values
data.dropna(inplace=True) #inplace means even when dropping the records containing null values the rest of the records stay at its place

#second approach: using the sklearn library to replace these missing values
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(independent[:,[3,4,7]])
independent[:,[3,4,7]]=imputer.transform(independent[:,[3,4,7]])
data.head()
```
