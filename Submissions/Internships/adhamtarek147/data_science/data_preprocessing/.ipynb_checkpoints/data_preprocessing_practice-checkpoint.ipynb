# importing libraries for data pre-processing

```python
import numpy as np
import pandas as pd
```

# loading dataset

```python
data=pd.read_csv("Employee.csv")
data.head()
```

# creating independent and dependent vector variables

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

# handling missing values

```python
#sum of missing data in the dataset
print(data.isnull().sum())

#first approach: handling the missing values by dropping the records that contain the missing values
data.dropna(inplace=True)#inplace means even when dropping the records containing null values the rest of the records stay at its place
data.head()
```

```python
#second approach: using the sklearn library to replace these missing values
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:,[3,4,7]])
x[:,[3,4,7]]=imputer.transform(x[:,[3,4,7]])
data.iloc[:,[3,4,7]]=x[:,[3,4,7]]
data.head()
```

# data encoding

```python
# one hot encoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[0])],remainder="passthrough")
x=np.array(ct.fit_transform(x))
print(x)
```

```python
#label encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=np.array(le.fit_transform(y))
print(y)
```

# Splitting data into Train and Test Set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# Feature Scaling | Standardization Vs Normalization

```python
# data.head()
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x[:,3:5]=scaler.fit_transform(x[:,3:5])
data.iloc[:,3:5]=x[:,3:5]
data.head()
```

# Outlier Detection & Removal

```python
import matplotlib.pyplot as plt
plt.hist(data["Age"],bins=15)
plt.show()
lowerLimit=data["Age"].quantile(0.5)
upperLimit=data["Age"].quantile(0.75)
data=data[(data["Age"]>lowerLimit) & (data["Age"]<upperLimit)]
data
```

# Data Mapping

```python
datamapping={
    "Male":1,
    "Female":2
}
data["Gender"]=data["Gender"].map(datamapping)
data.head()
```
