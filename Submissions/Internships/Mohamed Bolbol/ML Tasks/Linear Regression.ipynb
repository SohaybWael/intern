```python
df_copy["Gender"].replace({"Female" : 0, "Male" : 1}, inplace = True)
df_copy["Company Type"].replace({"Service" : 0, "Product" : 1}, inplace = True)
df_copy["WFH Setup Available"].replace({"No" : 0, "Yes" : 1}, inplace = True)
df_copy["Date of Joining"] = pd.to_datetime(df_copy["Date of Joining"])
df_copy["Join Year"] = df_copy["Date of Joining"].apply(lambda x : x.year)
df_copy["Join Month"] = df_copy["Date of Joining"].apply(lambda x : x.month)
df_copy["Join Day"] = df_copy["Date of Joining"].apply(lambda x : x.day)
df_copy.drop("Date of Joining", axis = 1, inplace = True)
df_copy.drop("Join Year", axis = 1, inplace = True)
df_copy.drop("Employee ID", axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split

# Assuming df_copy contains your DataFrame
# Extract features and target variable
X = df_copy[['Designation', 'Resource Allocation', 'Mental Fatigue Score', 'WFH Setup Available', 'Join Month', 'Join Day', 'Gender', 'Company Type']]
y = df_copy['Burn Rate']

# Get column names
feature_names = X.columns

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform feature selection Anova F-score
selector = SelectKBest(score_func=f_regression, k=5)  # Select top 5 features based on F-value
X_selected = selector.fit_transform(X_scaled, y)

# Get the selected feature names
selected_indices = selector.get_support(indices=True)
selected_feature_names = feature_names[selected_indices]

# Print the selected features
print("Selected Features:")
for feature in selected_feature_names:
    print("- ", feature)

# Add a bias term to the features
X_selected = np.c_[np.ones(X_selected.shape[0]), X_selected]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Initialize coefficients and bias
theta = np.zeros(X_train.shape[1])

# Define the linear regression equation
def predict(X, theta):
    return np.dot(X, theta)

# Define the gradient descent function
def gradient_descent(X, y, theta, learning_rate, iterations):
    m = len(y)
    
    for _ in range(iterations):
        error = predict(X, theta) - y
        gradient = (1/m) * np.dot(X.T, error)
        theta -= learning_rate * gradient
    
    return theta

# Perform gradient descent
learning_rate = 0.01
iterations = 1000
theta = gradient_descent(X_train, y_train, theta, learning_rate, iterations)

# Calculate R-squared on the test set
y_pred_test = predict(X_test, theta)
ss_total_test = np.sum((y_test - np.mean(y_test))**2)
ss_residual_test = np.sum((y_test - y_pred_test)**2)
r_squared_test = 1 - (ss_residual_test / ss_total_test)

print(f'R-squared value on test set: {r_squared_test}')
```
