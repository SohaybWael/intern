````python
import math
def perfectSquares(n):
    result = n/3
    int(math.floor(result))
    res = n-result
    if res % 2 == 0:
        if (res/2)%2 == 0:

            return(f"{result} + {res / 2} + {res/2} ")
        else:
            return(f"{result} + {res}")
    else:
        return(f"{result} + {res}")
````
