Write a function that takes in an array of integers and returns <u>the length of the longest peak in the array</u>.
</p>
<BR>

A peak is defined as adjacent integers in the array that are **strictly** increasing until they reach a tip (the highest value in the peak), at which point they become **strictly** decreasing. At least three integers are required to form a peak.

For example, the integers `1, 4, 10, 2` form a peak, but the integers `4, 0, 10` don't and neither do the integers `1, 2, 2, 0`. Similarly, the integers `1, 2, 3` don't form a peak because they aren't strictly decreasing after the `3`.

````python
def longestPeak(arr):
    ## write your solution here!
    n = len(arr)
    longest_peak_length = 0
    longest_peak_start = 0
    count = 0

    for i in range(1, n-1):# Our peak can only be from index 1 to n-2 index (n-1 is not included)(inclusive of lower bound, and exclusive of upper bound)
        if(arr[i]>arr[i-1] and arr[i]>arr[i+1]): # If arr[i] is greater than neighbors
            
            peak_length = 1 # nitializing count to 1 for the peak element itself
            
            j=i
            while j>0 and arr[j]>arr[j-1]: # Traversing the LHS of the peak until we reach the bottom of that mountain
                j-=1 # decrementing until we reach the bottom LHS
                peak_length+=1 # counting the steps until reaching bottom LHS


            k=i
            while k<n-1 and arr[k]>arr[k+1]: # Travering RHS of the mountain
                k+=1
                peak_length+=1 
            
            if peak_length > longest_peak_length:
                longest_peak_length = peak_length
                longest_peak_start = j # Start index of the longest peak


    
    return arr[longest_peak_start: longest_peak_start+longest_peak_length]

longest_peak = longestPeak([1, 2, 3, 3, 4, 4, -3, 2, 3, 0])
# print("Longest peak:", longest_peak)
# print("Length of longest peak:")
print(len(longest_peak))
````
